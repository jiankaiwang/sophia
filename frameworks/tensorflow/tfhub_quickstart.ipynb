# Tensorflow Hub

Reference: https://www.tensorflow.org/hub/

Tensorflow hub is a library for reusable machine learning modeules in tensorflow. A module is a self-contained piece of a Tensorflow graph, along with its weights, that can be reused across other, similar tasks. By reusing the module, a developer can train a model using a smaller dataset, improve generalization, or simply speed up the training.

Here we demo several topics using pretrained modules from tensorflow hub.

* Image Retraining
* Word Embedding
* Universal Sentence Encoder

# Word Embedding

````python
with tf.Graph().as_default():
    module_url = "https://tfhub.dev/google/nnlm-en-dim128-with-normalization/1"
    embed = hub.Module(module_url)
    embeddings = embed(["A long sentence.", "single-word"])

    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        sess.run(tf.tables_initializer())

        data = sess.run(embeddings)
        print(data)
````
